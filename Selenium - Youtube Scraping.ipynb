## Scarping Details of Youtube Videos

```python
import time
import selenium
import pandas as pd
from selenium import webdriver as wb
webD = wb.Chrome('chromedriver.exe')
webD.get('https://www.youtube.com/watch?v=wCoTJdhRcQE')
```

```python
name = webD.find_elements_by_xpath('//*[@id="container"]/h1/yt-formatted-string')[0].text
print(name)
views = webD.find_element_by_xpath('//*[@id="count"]/ytd-video-view-count-renderer/span[1]').text
print(views)
datepub = webD.find_element_by_xpath('//*[@id="info-strings"]/yt-formatted-string').text
print(datepub)
```

```python
name=[]
views=[]
timeadded=[]
links=[]
channels=[]
timelong=[]
for x in range(0,25):
    name.append(webD.find_elements_by_xpath('//*[@id="video-title"]')[x].text.capitalize())
    views.append(webD.find_elements_by_xpath('//*[@id="metadata-line"]/span[1]')[x].text[:-5])
    #subscripted to remove the word 'views'
    timeadded.append(webD.find_elements_by_xpath('//*[@id="metadata-line"]/span[2]')[x].text.replace('Streamed',''))
    links.append(webD.find_elements_by_xpath('//*[@id="thumbnail"]')[x].get_property('href'))
    channels.append(webD.find_elements_by_class_name('style-scope ytd-channel-name')[x+1].text)
    timelong.append(webD.find_elements_by_xpath('//*[@id="overlays"]/ytd-thumbnail-overlay-time-status-renderer/span')[x].text)
    
datayt = pd.DataFrame(
         {"NAME" : name,
          "VIEWS" : views,
          "DATE" : timeadded,
          "LINKS" : links,
          "CHANNELS" : channels,
          "LENGTH" : timelong
         })
datayt
```

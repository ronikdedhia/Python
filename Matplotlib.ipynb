# Matplotlib

Matplotlib is library of data visualization. It was created to try to replicate MatLab's plotting capabilities in Python. It is an excellent 2D and 3D graphics library for generating scientific figures. Matplotlib allows you to create reproducible figures programmatically. <br/>
http://matplotlib.org/ <br/>
https://matplotlib.org/2.0.2/gallery.html
<br/>
Some of the major Pros of Matplotlib are -<br/>
Generally easy to get started for simple plots<br/>
Support for custom labels and texts<br/>
Great control of every element in a figure<br/>
High-quality output in many formats<br/>
Very customizable in general<br/>

Data visualization allows us to quickly interpret the data and adjust different variables to see their effect

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

That line is only for jupyter notebooks, if you are using another editor, you'll use: plt.show() at the end of all your plotting commands to have the figure pop up in another window.

### Drawing Basic Charts

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

### Drawing Multiple Plots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # color options
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-'); # color options
# semicolon is to remove a line from output window
```

```python
plt.plot(x, y**2, 'b.-', label="y**2", alpha=0.5)
plt.plot(x, y**3, 'r--', label="y**3")
plt.xlabel('Set X Label')
plt.ylabel('Set y Label')
plt.title('Set Title')
plt.legend(loc=0); # ranges from 0-4, 0=optimal, 1-4=corners
```

### Drawing Figure with Multiple Axes

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50);
```

### Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
# line width and line style

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z');
```

```python
fig, ax = plt.subplots(2, 3)
fig.tight_layout()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
```

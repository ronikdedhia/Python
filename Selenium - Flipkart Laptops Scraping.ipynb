## Scraping Details of Laptops listed on E-Commerce Website

```python
import selenium
import time
from selenium import webdriver as wb
import pandas as pd
webD = wb.Chrome('chromedriver.exe')
webD.get('https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=385d9c97-16c3-4901-aef1-c25c8bc2f759&as-backfill=on')
```

```python
names=[]
ratings=[]
noofratings=[]
noofreviews=[]
features=[]
costs=[]
for i in range(0,24):
    names.append(webD.find_elements_by_class_name('_4rR01T')[i].text.split('-')[0].strip())
    ratings.append(webD.find_elements_by_class_name('_3LWZlK')[i].text)
    features.append(webD.find_elements_by_class_name('fMghEO')[i].text)
    costs.append(webD.find_elements_by_class_name('_25b18c')[i].text.split('\n')[0].strip())
    if i == 23:
        next_url = webD.find_element_by_class_name('_1LKTO3').get_property('href')
        webD.get(next_url)
        time.sleep(10)
        
for i in range(0,24):
    names.append(webD.find_elements_by_class_name('_4rR01T')[i].text.split('-')[0].strip())
    ratings.append(webD.find_elements_by_class_name('_3LWZlK')[i].text)
    features.append(webD.find_elements_by_class_name('fMghEO')[i].text)
    costs.append(webD.find_elements_by_class_name('_25b18c')[i].text.split('\n')[0].strip())
```

```python
datayt = pd.DataFrame(
         {"NAME" : names,
          "RATINGS" : ratings,
          "FEATURES" : features,
          "COST" : costs,
         })
datayt
```
